```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Datasets of Airbnb in Singapore
filename = "sgairbnb.csv"
neighbourhood_filename = "neighbourhoods_utf8.csv"
airbnb_df = pd.read_csv(filename)
airbnb_df
```

```python
# Datasets of neighbourhoods' names in Singapore
neighbourhood_filename = 'neighbourhoods_utf8.csv'
neighbourhood = pd.read_csv(neighbourhood_filename)
neighbourhood.head()
```

```python
neighbourhood_list = list(neighbourhood['neighbourhood'])
print(neighbourhood_list)
```

```python
# Master plan map of Singapore
file_path = "./master-plan-2014-planning-area-boundary-web-shp/MP14_PLNG_AREA_WEB_PL.shp" # GeoDataFrame type
sg_map = gpd.read_file(file_path)
```

```python
# Create a new column which represents the center of each neighbourhood
sg_map['center'] = sg_map['geometry'].centroid
sg_map
```

```python
# Create a list of neighbourhoods with their respective area in map and center of the area
neighbourhood_loc = []

for i in range(len(sg_map)):
    neighbourhood_loc.append((sg_map['PLN_AREA_N'][i].lower(), sg_map['geometry'][i], sg_map['center'][i]))

print(neighbourhood_loc)
```

```python
# Calculate the total number of Airbnb accommodations in each neighbourhood
neighbourhood_dict = {}

airbnb_neighbourhood = list(airbnb_df['neighbourhood'])
for line in airbnb_neighbourhood:
    if line not in neighbourhood_dict:
        neighbourhood_dict[line] = 1
    elif line in neighbourhood_dict:
        neighbourhood_dict[line] += 1

print(neighbourhood_dict)
```

```python
# Updating the list of neighbourhoods into a dictionary
for k, v in neighbourhood_dict.items():
    for loc in neighbourhood_loc:
        if k.lower() == loc[0]:
            neighbourhood_dict[k] = [k, v, loc[1], loc[2]]

print(neighbourhood_dict)
```

```python
# Convert the dictionary of neighbourhoods to a dataframe
neighbourhood_df = pd.DataFrame.from_dict(neighbourhood_dict, orient='index', columns=['neighbourhood', 'num_of_listings', 'geometry', 'center'])
print(neighbourhood_df['geometry'])
```

```python
# Convert the neighbourhood dataframe into a geodataframe for geospatial visualisation
neighbourhood_gdf = gpd.GeoDataFrame(neighbourhood_df)

neighbourhood_gdf_copy = neighbourhood_gdf.copy()
neighbourhood_gdf_copy.set_geometry('center', inplace=True)
```

```python
# Visualise the neighbourhood geodataframe
# Density of colours varies according to the number of Airbnb listings in each neighbourhood
neighbourhood_gdf.plot(column='num_of_listings', cmap='PiYG', figsize=(25, 25), legend=True)

for x, y, label in zip(neighbourhood_gdf_copy.center.x, neighbourhood_gdf_copy.center.y, neighbourhood_gdf['neighbourhood']):
    plt.text(x, y, label, fontsize=12, color='black')
```
